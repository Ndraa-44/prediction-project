## Deskripsi Informasi : 
1. jumlah kursus yang diikuti
2. rata-rata waktu belajar tiap hari 
3. nilai yang diperoleh

problem : prediksi nilai siswa berdasarkan no 1 dan 2

```python
import pandas as pd
import numpy as np
data = pd.read_csv('nilai_pelajar.csv')
print(data.head(10))
```

Ternyata hanya ada 3 kolom/fitur pada dataset

```python
# Mengecek apakah ada data kosong
print(data.isnull().sum())
```

```python
data["number_courses"].value_counts()
```

```python
# Kita mulai membuat grafik plot
import plotly.express as px
import plotly.io as pio
from sklearn.model_selection import train_test_split
```

```python
figure = px.scatter(data_frame=data, x="number_courses", y="Marks", size="time_study",
                    title="Jumlah Kursus dan Nilai")
figure.show()
```

```python
# Ternyata banyak sedikitnya kursus yang diambil tidak terlalu berpengaruh pada nilai.
# Sekarang coba kita plotting antara jam belajar dengan nilai.

figure = px.scatter(data_frame=data, x="time_study", y="Marks", size="number_courses",
                    title="Jam Belajar dan Nilai")
figure.show()
```

```python
# Ternyata ada hubungan linear antara jam belajar dengan nilai.
correlation = data.corr()
print(correlation["Marks"].sort_values(ascending=False))
```

```python
# jadi memang korelasi yang paling besar adalah antara jam belajar dengan nilai.
# Membuat model prediksi sederhana menggunakan Linear Regression
from sklearn.linear_model import LinearRegression
x = np.array(data[["time_study", "number_courses"]])
y = np.array(data["Marks"])
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=41)
```

```python
# Fitting model
model = LinearRegression()
model.fit(x_train, y_train)
model.score(x_test, y_test)
```

```python
# Mari kita prediksi nilai
# fitur yang kita input adalah jam belajar dan jumlah kursus
features = np.array([[9.567, 7]])
model.predict(features)
```
